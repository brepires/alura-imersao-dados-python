```python
column_translation = {
    'ano_trabalho': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_funcionario': 'residencia',
    'taxa_remoto': 'remoto',
    'local_empresa': 'local_empresa',
    'tamanho_empresa': 'tamanho_empresa'
}

df.rename(columns=column_translation, inplace=True)
print(df.columns)
```

```python
df['senioridade'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Medio',
    'EN': 'Junior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
print(df['senioridade'].value_counts())
```

```python
contrato_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelance'
}
df['contrato'] = df['contrato'].replace(contrato_translation)
print(df['contrato'].value_counts())
```

```python
tamanho_empresa_translation = {
    'M': 'Media',
    'L': 'Grande',
    'S': 'Pequena'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_translation)
print(df['tamanho_empresa'].value_counts())
```

```python
remoto_translation = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(remoto_translation)
print(df['remoto'].value_counts())
```
